## Overview

This project is designed to find the key factors that determine the price of housing sales, in order to determine what to look for when considering investing in housing. Within we analyse the data of the sales of housing within Washington's King County. In the following I will show a brief overview of the techniques used to determine the factors in question.

## Business Problem

When investing in housing, it's helful to know the factors that lead to price incereases. This project is aimed at exactly that, determining which qualities to look for to get the maximum return on your interest.

## Data Understanding

The data used in this project comes from the King's County House Dataset, containing over 20,000 house sales over the course of 2014-2015. It contains a wealth of information on the houses sold, from the square footage to the number of bathrooms. These are compared to the price of sale to determine which factors contributed the most to prices.

## Data Preparation

After cleaning the data, the first step was to determine which factors were linked with each other to avoid multicolinearity from muddling the results, and also which factors were linked with the price in order to narrow the scope of our study.

Before taking this to multiple regression, which would show us our factors, we had to demonstrate normality and homoscedacity in our data. After a quick analysis we determined that several of the factors, namely the price and the various square areas, had to be transformed to their natural log.

For our final step in data preparation we removed outliers in our price-correlated features and proceeded to multiple regression.

## Data Modeling

The below data shows that of our factors, the most powerful influences on housing prices according to our data are the square footage of the living area, the grade of the house (a rating determined by the King County rating system), and the square footage of the living area of the closest 15 neighbors.

There is one more factor that has a notable influence named "view", but unfortunately I did not notice until writing this that it doesn't actually refer to the view from the house but the far more nebulous "has been viewed", so for the purposes of this study I have decided to ignore it.

As a final round of validation we found the mean squared error of our variables with multiple cross analysis. Our average MSE was 0.117, showing a relatively small prediction error. This means our model is very reliable.

## Conclusions

Our study shows that when considering investing in housing property, you should look for areas with houses that have large living spaces, not just the living space of the property itself. It also shows that the King County grading system is reliable at predicting price of the property, this may or may not be generalisable to other grading systems, further studies would be needed in this area.
